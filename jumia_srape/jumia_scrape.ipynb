```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import pandas as pd 
```

```python
url = 'https://www.jumia.com.ng/catalog/?q=inverter'
hd = UserAgent()
userAgent = hd.random
headers = {'User-Agent': userAgent}

page = requests.get(url, headers = headers)
soup = BeautifulSoup(page.content,'html.parser')
```

```python
inverter_list = soup.find_all('article', class_ = 'prd _fb col c-prd')

names = []
prices = []
star_ratings = []

for list in inverter_list:

    #names
    if list.find('h3', class_ = 'name') is not None:
        name = list.find('h3', class_ = 'name')
        names.append(name.text)
    else:
        names.append('None')

    #prices
    if list.find('div', class_ = 'prc') is not None:
        price = list.find('div', class_ = 'prc')
        prices.append(price.text)
    else:
        prices.append('None')

    #star_rating
    if list.find('div', class_ = 'stars _s') is not None:
        star_vote = list.find('div', class_ = 'stars _s')
        star_ratings.append(star_vote.text)
    else:
        star_ratings.append('None')



inverter_df = pd.DataFrame({
    'name':names,
    'prices': prices,
    'rating': star_ratings
})
inverter_df
```

```python
# Saving data to a csv file
inverter_df.to_csv('inverter_data.csv')
```
